# Data Visualization with Matplotlib and Seaborn

###### 1. Creating Figures and Subplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Ploting ...
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(2,2,1) #(2,2,1):two rows and two columns and 1: first plot
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot([1,2,3,4,5])
ax2.plot(np.random.randn(20).cumsum())
```

```python
# subplots ...
fig, ax = plt.subplots(3,
                       3, 
                       figsize=(8,6))
ax[0,0].plot([1,2,3,4,5])
ax[2,2].plot(np.random.randn(20).cumsum())

# organize the plots ...
fig.subplots_adjust(hspace=.4, wspace=0.4)
```

###### 3. Customizing Ticks and Labels

```python
data = np.random.randn(200).cumsum()

# Plot ...
fig, ax = plt.subplots(1,1)
ax.plot(data)

# ticks = ax.set_xticks(range(0,200,50))
ticks = ax.set_xticks([0,50,100,150,200])
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=60)

# Adding the labels ...
ax.set_xlabel('Cumulative sum')
ax.set_ylabel('Values')

# Addin a title ...
ax.set_title('Cumulative sum for Random Numbers')
```

###### 4. Adding Legends

```python
# Example ...
data1 = np.random.randn(200).cumsum()
data2 = np.random.randn(200).cumsum()
data3 = np.random.randn(200).cumsum()

# Plot ...
fig, ax = plt.subplots(1,1)
ax.plot(data1, label = 'data1')
ax.plot(data2, label = 'data2')
ax.plot(data3, label = 'data3')

# Adding the legend ...
ax.legend(loc='best')

# ticks = ax.set_xticks(range(0,200,50))
ticks = ax.set_xticks([0,50,100,150,200])
labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=60)

# Adding the labels ...
ax.set_xlabel('Cumulative sum')
ax.set_ylabel('Values')

# Addin a title ...
ax.set_title('Cumulative sum for Random Numbers')

```

###### 5. Adding Texts and Arrows on a Plot

```python
# Two methods:
                # Texts  => text()
                # Arrows => arrow()
```

```python
# Example ...
x = np.arange(0,10,0.1)
y = np.sin(x)

# Plot ...
fig, ax = plt.subplots(1,1)
ax.plot(x,y)

# Add text ...
ax.text(4,0.7,'y=sin(x)', fontsize=13)

# Add the text in the middle ...
#ax.text(0.5,0.5,'y=sin(x)', fontsize=13, transform=ax.transAxes)


# Adding arrows ...
ax.arrow(0.5,1,0.6,0, width=0.05)
ax.arrow(6.8,1,0.6,0, width=0.05)
```

```python
# Adding text() and arrow() together ...
# Plot ...
fig, ax = plt.subplots(1,1)
ax.plot(x,y)

# Add text ...
ax.text(0.35,0.8,'Wave length', fontsize=13, transform=ax.transAxes)


# Adding arrows ...
ax.arrow(1.8,1,2.5,0, width=0.05)
ax.arrow(7.8,1,-2.5,0, width=0.05)
```

###### 6. Adding Annotations and Drawings on a Plot

```python
# Annotations is used in figures to draw attention to special features or important points on the graph.
```

```python
# Example ...

x = np.arange(6)
y = np.array([23,34,65,78,51,55])

# Plot ...
fig, ax = plt.subplots(1,1)
ax.plot(x,y, 'ko--')

# Adding an argument to the graph ...
ax.annotate('Maximum', 
            xy=(x[3],y[3]),
            xytext=(3,50),
            arrowprops = dict(facecolor='blue',shrink=0.1))

ax.annotate('Minimum', 
            xy=(x[0],y[0]),
            xytext=(0,50),
            arrowprops = dict(facecolor='green',shrink=0.1))
```

```python
# Adding a shape 'circle'...
fig, ax = plt.subplots(1,1)
circle = plt.Circle((0.5,0.5),0.2, fc='y')
ax.add_patch(circle)
```

```python
# Adding a shape 'rectangle' ...
fig, ax = plt.subplots(1,1)
rect = plt.Rectangle((0.3,0.5),0.5,0.2, color='y', alpha=0.5)
ax.add_patch(rect)
```

```python
# Adding a shape 'Polygon' ...
fig, ax = plt.subplots(1,1)
tri = plt.Polygon([[0.25,0.25],[0.55,0.5],[0.4,0.7]], color='g', alpha=0.5)
ax.add_patch(tri)
```

###### 11. Histograms and Density Plots

```python
# How to plot 'histograms' and 'density plots'.
# A histogram: displays the frequency of values.
# The data is split into discrete ranges(intervals), and each
# bar represents how many data points fall in that range.

# A histogram: is used as a representation of the distribution
# of the data, as it displays the shape and the spread of continuous data.
```

```python
# Example ...
df = pd.read_csv('data/ex13.csv')
df.head()
```

```python
# Plot ...
df['age'].plot.hist(bins=13,
                    color='g', 
                    edgecolor='black', 
                    linewidth=1)
```

```python
# Plot ...
df['wage'].plot.hist(bins=20,
                    color='g', 
                    edgecolor='black', 
                    linewidth=1)
```

```python
# Density plot is used to visualize the distribution of the data.
# Its advantage is that the density plots are not affected by the number of bins.
```

```python
# Density plot ...
df['wage'].plot.density()
```

```python
# plot the boths of them
sns.displot(df['wage'], bins=12, color='r')
```

```python
# plot the boths of them
sns.displot(df['wage'], kind='kde', color='b')
```

###### 13. Factor Plots for Categorical Data

```python
# Factor plots are used to visualize categorical data.
# catplot()
```

```python
# Example ...
credit = pd.read_csv('data/credit.csv')
credit.head()
```

```python
# Plot ...
sns.catplot(x='Gender',
            y='Income',
            data=credit)
```

```python
# Plot using kind='bar'...
sns.catplot(x='Gender',
            y='Income',
            data=credit,
            kind='bar')
```

```python
# Plot using kind='bar'...
sns.catplot(x='Gender',
            y='Income',
            data=credit,
            kind='bar',
            ci=False)
```

```python
# Plot using kind='box'...
sns.catplot(x='Gender',
            y='Income',
            data=credit,
            kind='box')

# The two lines represent the 'min income' and the 'max income'
# The box represent 'the range of 50% of cases'
                    # => means: 50% of men receive an income belween 20 and 65 thousand.
                    # => means: 50% of women receive an income belween 20 and 40 thousand.
```

```python
# Plot ...
sns.catplot(x='Ethnicity', 
            y='Income', 
            hue='Gender', 
            data=credit, 
            kind = 'box')
```
