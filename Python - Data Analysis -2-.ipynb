###### 2. Reordering and Sorting Index Levels

```python
# How you can 'sort' and 'reorder' the levels in the multi-indexed DF
```

```python
# Example ...
data = pd.DataFrame(np.random.randint(100,size=(6,4)),
                   index = [['green','green','black','black','yellow','yellow'],
                            ['one','two','one','two','one','two']])

data
```

```python
# reorder the position of levels (swaplevel()) level 1 <===> level 2
# Step-1-
data2 = data.swaplevel(0,1)
data2
```

```python
# Step-2-
data2.sort_index(level=0) # level=0 => the first index level will be sorted.
```

```python
# Step-1- and Step-2- together ...
data.swaplevel(0,1).sort_index(level=0)
```

# 2. Data Wrangling -2- 

###### 1. Merging Datasets on Keys (common columns)

```python
# How to merge and combine datasets.
# the 'merge()' method is used to merge two dataframes based on a 'common column'.
```

```python
# Example 1 ...
data1 = pd.DataFrame({'states':['California', 'Georgia', 'Florida', 'Arizona'],
                     'Population':[40,10,21,7]})

data1
```

```python
# Example 2 ...
data2 = pd.DataFrame({'states':['Arizona', 'Colorado', 'Indiana', 'Florida'],
                     'Area':[113,104,36,65]})

data2
```

```python
# Merge the both dataset(intersection merging methode) using => merge()
pd.merge(data1,data2, on='states') #pd.merge(data1,data2)
```

```python
# Merge the both dataset => "how='outer'"
pd.merge(data1,data2, how='outer')
```

```python
# Merge the both dataset => "how='left '"
pd.merge(data1,data2, how='left')
```

```python
# Merge the both dataset => "how='right '"
pd.merge(data1,data2, how='right')
```

```python
# Another example contain two different columns names ...
data3 = pd.DataFrame({'states_name':['California', 'Georgia', 'Florida', 'Arizona'],
                     'Population':[40,10,21,7]})

data3
```

```python
#
data2
```

```python
# contain two different names
pd.merge(data3,data2,left_on='states_name',right_on='states')
```

```python
# Example ...
data4 = pd.DataFrame({'states':['California', 'Georgia', 'Florida', 'Arizona'],
                     'Population':[40,10,21,7],
                     'water':[4,6,7,2]})

data4
```

```python
#
data5 = pd.DataFrame({'states':['Arizona', 'Colorado', 'Indiana', 'Florida'],
                     'Area':[113,104,36,65],
                     'water':[2,8,3,7]})

data5
```

```python
# merge the dataset4 and dataset5 ...
pd.merge(data4,
         data5, 
         on='states',
         how='outer')
```

```python
# rename them to know easily the data for each column ...
# ===> using 'suffixes()'

pd.merge(data4,
         data5, 
         on='states',
         how='outer',
         suffixes = ('_data4', '_data5'))
```

##### 3. Concatenating Along an Axis 

```python
# How to concatenate DF along an axis.
# Concatenate DF is widely used in DS and Data Analysis.
```

```python
# Example ...
ser1 = pd.Series([1,2,3],
                 index = ['a','b','c'])
ser1
```

# 3. Data Wrangling -3- 

###### 1. Reshaping by Stacking and Unstacking

```python
# reshaping: transforming the structure of the data to make it appropriate for further analysis.
# Reshaping includes:
                    # Stacking.
                    # Melting.
                    # Pivoting.
            
# We will focus on the stack() and unstack() methods.
# ===========
# stacking: is moving the intermost column index to become the intermost row index.
# unstacking: is moving the intermost row index to become the intermost column index.
```

```python
# Example ...

data = pd.read_csv('data/ex6.csv', index_col = 'branch number')
data
```

```python
# Covert the columns to rows ...
data2 = data.stack()
data2
```

```python
# reconvert the rows to columns (transform it to its original mode)...
data2.unstack()
```

```python
# transform the rows to columns ...
data2.unstack(0) # '0' refers to the outer index (here it is the branch number)
```

```python
# Method-2- transform the rows to columns ...
data2.unstack('branch number') 
```

```python
# Example -2- 'double index'...

frame1 = pd.read_csv('data/ex5.csv', index_col = ['year','quarters'])
frame1
```

```python
# unstack() method ...
frame1.unstack()
```

```python
# set 'year' as index
frame2 = frame1.unstack('year')
frame2
```

```python
# Back to its original ...
frame2.stack()
```

###### 2. Reshaping by Melting (Wide to Long )

```python
# 'melt()' method: is used to transforms multiple columns into a single column.
# It's using for reshaping DF.
# met() method is the oposite of pivot() method.
```

```python
# Example ...
data = pd.read_csv('data/ex7.csv')
data
```

```python
# melt the data together ...
data2 = data.melt(id_vars='year')
data2
```

```python
# Rename the columns ...
data2 = data.melt(id_vars='year',
                   var_name='sales',
                   value_name ='amount')
data2
```

```python
# Example ...

data3 = pd.read_csv('data/ex8.csv', index_col='date')
data3
```

```python
# We assume to want google and apple columns transform to one column.
data3.melt(var_name='Company',
           value_name='closing price')
```

```python
#
data4 = data3.melt(var_name='Company',
                   value_name='closing price',
                   ignore_index=False)
data4
```

```python
# aranging the index ...
data4.sort_index()
```
